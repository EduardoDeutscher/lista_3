```python
#1. Salve a base de dados escolhida no seu drive e importe para um notebook do Colab, criando um Dataframe
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/class_diabetes.csv")
dataDF=pd.DataFrame(data)
display(dataDF)
```

```python
#2.Inspecione as colunas do conjunto de dados (Nome das colunas e tipo de dados)
display(data.info())
```

```python
#3.Escolha uma coluna com valores numÃ©ricos e apague todas as linhas que possuam valor nulo
dataDF.dropna(subset=['preg'],inplace=True)
display(dataDF)
```

```python
#4.Calcule a mÃ©dia desta mesma coluna
media = dataDF['preg'].mean()
print(media)
```

```python
#5.Escolha um critÃ©rio para eliminar as linhas que possuam valores muito distantes da mÃ©dia e crie um novo Dataframe.
linremov1 = dataDF[(dataDF['preg']>=4.2)].index
linremov2 = dataDF[(dataDF['preg']<3.4)].index
dataDF_remove1 = dataDF.drop(linremov1,axis=0)
dataDf_remove2 = dataDF_remove1.drop(linremov2,axis=0)
display(dataDf_remove2)
```

```python
#6.Crie um novo Dataframe que contenha apenas 5 colunas do conjunto anterior.
col1 = dataDF['preg']
col2 = dataDF['plas']
col3 = dataDF['pres']
col4 = dataDF['skin']
col5 = dataDF['insu']
novoDF = pd.concat([col1,col2,col3,col4,col5],axis = 1)
display(novoDF)
```

```python
#7.Escolha duas colunas numÃ©ricas e crie uma nova coluna que siga a regra: (Coluna A - Coluna B)*nÃºmero aleatÃ³rio (entre 0 e 1).
import random
num=random.random()
print(num)
novoDF['nova'] =(novoDF['preg'] + novoDF['plas'])*num
display(novoDF)
```

```python
#8.Embaralhe aleatoriamente o dataframe (Ã­ndices e valores).
import numpy as np
DF3 = novoDF.reindex(np.random.permutation(novoDF.index))
display(DF3)
```

```python
#9.Ordene o Dataframe seguindo um critÃ©rio definido por vocÃª (Explique o critÃ©rio)
#CritÃ©rio: valores de preg em ordem decrescente
ordem = DF3.sort_values(by = 'preg',ascending = False)
display(ordem)
```

```python
#10.Escolha uma coluna numÃ©rica e um coluna com strings e utilize a funÃ§Ã£o Groupby para mostrar os totais da  coluna.
#NÃ£o possuo nenhum valor do tipo string
agrupamento = DF3.groupby(['preg']).count()
display(agrupamento)
```
